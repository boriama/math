# Тема "Множество"

1. Даны три множества a,b и с. Необходимо выполнить все изученные виды бинарных операций над всеми комбинациями множеств.


(Буду сразу приводить решение на pythone)

```python
a = set([1, 2, 3, 4])
b = set([3, 4, 5, 6])
c = set([6, 7, 2, 4])
```

Пересечение множеств:

```python
set.union(a, b, c)
```

Объединение множеств:

```python
set.intersection(a, b, c)
```

Разность:

```python
 a.difference(b).difference(c)
```

Симметрическая разность:

```python
set.union(
     a.difference(b).difference(c),
     b.difference(c).difference(a),
     c.difference(a).difference(b)
)
```

Декартово произведение:

```python
from itertools import product
list(product(a, b, c))
```

# Тема 3 “Последовательность”

1. Даны 4 последовательности. Необходимо:

    a.исследовать их на монотонность; 
    
    b.исследовать на ограниченность;
    
    c.найти пятый по счету член.


a) "\{a_n\}^\infty_{n=1} = 2^n -n"

Для определения монотонности иследуем разность "a_n - a_{n+1}":
"a_n - a_{n+1} = 2^n -n - 2^{n+1} +n + 1 = 2^n(1 - 2) +1 = 1 - 2^n < 0, \forall n \in \mathbb{N} => a_n < a_{n+1}"
Следовательно, последовательность является монотонно возрастающей

Исследуем ограниченность:
"\lim_{n \to \infty } 2^n -n \to \infty ", значит последовательность неограниченна

"a_5 = 2^5 - 5 = 27"

b) "\{b_n\}^\infty_{n=2} = \frac{1}{1-n}"

Для определения монотонности иследуем разность "b_n - b_{n+1}":
"b_n - b_{n+1} = \frac{1}{1-n} - \frac{1}{1-n + 1} = \frac{1}{1-n} - \frac{1}{-n} = \frac{-n\ -\ (1-n)}{(1-n)(-n)} = - \frac{1}{n^2 -n} < 0, \forall n \in \mathbb{N} => b_n < b_{n+1}"
Следовательно, последовательность является монотонно возрастающей

Исследуем ограниченность:
"\lim_{n \to \infty } \frac{1}{1-n} \to 0", значит последовательность ограниченна

"b_5 = \frac{1}{1-5} = - \frac{1}{4}"

c) "\{c_n\}^\infty_{n=1} = -1^n + \sqrt{2n}"

Для определения монотонности иследуем разность "с_n - с_{n+1}":
"c_n - c_{n+1} = -1^n + \sqrt{2n} - (-1)^{n+1} - \sqrt{2(n+1)} = \sqrt{2}(\sqrt{n}-\sqrt{n+1}) < 0 , \forall n \in \mathbb{N} => c_n < c_{n+1}"
Следовательно, последовательность является монотонно возрастающей

Исследуем ограниченность:
"\lim_{n \to \infty } -1^n + \sqrt{2n} = \lim_{n \to \infty } \sqrt{2n} -1 \to \infty" , значит последовательность неограниченна

"c_5 = -1 + \sqrt{2\cdot5} = \sqrt{10} -1 \approx 2.162"

d) "\{d_n\}^\infty_{n=1} = (-1)^{2n} + \frac{1}{n^2}"

Для определения монотонности иследуем разность "d_n - d_{n+1}":
"d_n - d_{n+1} = (-1)^{2n} + \frac1{n^2} - (-1)^{2(n+1)} + \frac1{(n+1)^2} = \frac{1}{n^2} - \frac{1}{(n+1)^2} > 0, \forall n \in \mathbb{N} => d_n > d_{n+1}"
Следовательно, последовательность является монотонно убывающей

Исследуем ограниченность:
"\lim_{n \to \infty } (-1)^{2n} + \frac{1}{n^2} = \pm1", то есть, имеем ограничение и сверху и снизу, значит последовательность ограниченна

"d_5 = (-1)^{2\cdot5} + \frac{1}{5^2} = 1 + \frac{1}{25} = 1.04"
